### Potencial Gravitacional 

```python
#def funciones
def Epm(G,m1,m2,r):
    return - G*m1*m2/r
def Ep(r):
    return Epm(1,1,1,r)
def Ektl(L,m2,r):
    return L**2/(2*m2*r**2)
def Ek(L,r):
    return Ektl(L,1,r)
```

```python

rinicial=1
rfinal=50
rpasos=rfinal-rinicial+1
#tlin=np.linspace(0,50,51)
rlin=np.linspace(rinicial,rfinal,rpasos)
xlinEp=Ep(rlin/5)
xlinEk1=Ek(1,rlin/5)
xlinEk12=Ek(1.2,rlin/5)
xlinEk15=Ek(1.5,rlin/5)
xlinEk2=Ek(2,rlin/5)
#xzero=zero(rlin)
```

```python
plt.title("Ep,Ek")
plt.plot(rlin,xlinEp,label="Ep")
plt.plot(rlin,xlinEk2,label="Ek, L=2")
#plt.plot(rlin,xlinEk1,label="Ek, L=1")
plt.plot(rlin,xlinEp+xlinEk2,label="Ep+Ek")
plt.plot(rlin,xlinEp+xlinEk15,label="Ep+Ek, L=1.5")
plt.plot(rlin,xlinEp+xlinEk12,label="Ep+Ek, L=1.2")
plt.plot(rlin,xlinEp+xlinEk1,label="Ep+Ek, L=1")
plt.axvline(x=5,color="gray",linestyle="--")
#plt.axvline(x=5,color="gray")
#plt.axvline(x=5,color="gray")
#plt.axvline(x=11.5,color="gray")
plt.legend()

plt.xlim(0,50)
plt.ylim(-1,0.4)
plt.grid()
#plt.show()
```
